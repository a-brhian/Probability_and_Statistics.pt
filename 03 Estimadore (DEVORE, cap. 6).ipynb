##Bernoulli

A variável aleatória (va) *Gender* é do tipo binária, logo podemos atribuir a ela uma fmp do tipo ***Bernoulli(p)***. Considere que *p* é a proporção de encontrar *homens* nesse banco de dados, então podemos estimar *p* por meio do estimador $\hat{p}$ escrito nos seguintes códigos:

```r
p.tot <- table(dados.f$Gender)/dim(dados.f)[1]; p.tot
cat('\n')
p <- round(p.tot[2],3); p
```

Logo, o banco de dados que estamos trabalhando tem $9.7%$ de usuários masculinos.

## Poisson

Se consideramos que a va *Conflicts_Over_Social_Media* tem uma distribuição $Poisson(\lambda)$, então podemos responder perguntas como: *Qual a probabilidade de uma pessoa ter exatamente 4 conflitos por conta da rede social?*

O parâmetro $\lambda$ pode ser estimado por meio do seguinte algoritmo:

```r
lambda <- round(mean(dados.f$Conflicts_Over_Social_Media),3); cat(lambda,"conflitos")
```

A probabilidade de uma pessoa ter exatamente 4 conflitos pode ser calculado por meio da seguinte função:

```r
cat(round(dpois(x=4,lambda),3)*100,"%")
```

Agora, vamos construir um histograma da variável `Conflicts_Over_Social_Media` e sobrepor uma fmp *Poisson(16.1)*:

```r
# Criar histograma
hist(dados.f$Conflicts_Over_Social_Media,
     breaks = seq(-0.5, max(dados.f$Conflicts_Over_Social_Media)+0.5, by = 1),
     freq = FALSE,
     col = "lightblue",
     main = "Histograma com Distribuição de Poisson",
     xlab = "Conflicts Over Social Media",
     ylab = "Frequência")

# Gerar os valores da Poisson
x_vals <- 0:max(dados.f$Conflicts_Over_Social_Media)
pois_probs <- dpois(x_vals, lambda = lambda)

# Como o histograma está em densidade, dividir pela largura dos intervalos (que é 1, então não precisa ajustar)
# Sobrepor a distribuição de Poisson
lines(x_vals, pois_probs, type = "h", lwd = 2, col = "darkorange")
points(x_vals, pois_probs, pch = 16, col = "darkorange")

# Legenda
legend("topright", legend = c("Histograma", paste0("Poisson(", lambda, ")")),
       col = c("lightblue", "darkorange"), lty = c(NA, 1), pch = c(15, 16))
```

O gráfico mostra que a fmp *Poisson(2.884)* não se ajustou tão bem a fdp da variável `Conflicts_Over_Social_Media`, pois atribuiu probabilidade para valores de conflitos inexistentes na amostra.

## Gama

Vamos analisar o histograma da va *Sleep_Hours_Per_Night*:

```r
x <- dados.f$Sleep_Hours_Per_Night

hist(x,
     breaks = 20,
     freq = FALSE,
     col = "lightblue",
     border = "white",
     main = "Ajuste da Distribuição Gamma a 'Sleep Hours per Night'",
     xlab = "Sleep Hours per Night",
     ylab = "Frequência")
```

Nota-se que essa va só pode apresentar valores positivos além de apresentar uma leve inclinação para direita.

Para esse tipo de dado, uma distribuição que consegue cobrir esse tipo de característica é a frdp *Gama($\alpha$,$\beta$)*.

Esses parâmetros podem ser calculados por um método chamado de "Método da Máxima Verossimilhança", estimados pelos seguintes códigos:

```r
require("MASS")

suppressWarnings(fitdistr(dados.f$Sleep_Hours_Per_Night, "gamma"))
```

Os parâmetros também podem ser calculados por meio de um outro método denominado de "Método dos Momentos", a saber:

```r
alpha.p <- mean(dados.f$Sleep_Hours_Per_Night)^2/sd(dados.f$Sleep_Hours_Per_Night); round(alpha.p,3)
```

```r
beta.p <- sd(dados.f$Sleep_Hours_Per_Night)/mean(dados.f$Sleep_Hours_Per_Night); round(beta.p,3)
```

O ajuste do histograma da variável com os parâmetros da fdp Gama estimado pelos dois métodos podem ser visualizados no gráfico a seguir:

```r
# 1. Estimativa dos parâmetros pelos Momentos
media_x <- mean(x)
var_x <- var(x)

shape_mom <- (media_x^2) / var_x
rate_mom <- media_x / var_x

# 2. Estimativa pela Máxima Verossimilhança
fit_gamma <- suppressWarnings(fitdistr(x, "gamma"))

shape_mle <- fit_gamma$estimate['shape']
rate_mle <- fit_gamma$estimate['rate']

# 3. Criar histograma da variável
hist(x,
     breaks = 20,
     xlim = c(1,11),
     freq = FALSE,
     col = "lightblue",
     border = "white",
     main = "Ajuste da Distribuição Gamma a 'Sleep Hours per Night'",
     xlab = "Sleep Hours per Night",
     ylab = "Frequência/Densidade de Frequência")

# 4. Sobrepor as densidades estimadas

# Sequência para traçar as densidades
x_seq <- seq(min(x)-1, max(x)+2, length.out = 500)

# Densidade Gamma pelos Momentos
dens_mom <- dgamma(x_seq, shape = shape_mom, rate = rate_mom)

# Densidade Gamma pela Verossimilhança
dens_mle <- dgamma(x_seq, shape = shape_mle, rate = rate_mle)

# Linhas de densidade
lines(x_seq, dens_mom, col = "blue", lwd = 4, lty = 2)
lines(x_seq, dens_mle, col = adjustcolor("darkorange",.75), lwd = 4, lty = 1)

# 5. Legenda
legend("topleft",
       legend = c("Gamma - Momentos", "Gamma - Verossimilhança"),
       col = c("blue", "darkorange"),
       lwd = 4,
       lty = c(2, 1),
       bty = "n")
```

Apesar dos parâmetros serem diferentes, visualmente os dois métodos parecem estimar duas fdps Gama que se ajustam bem aos dados.

Uma forma de decidir qual a melhor fdp Gama é por meio de um teste não-paramétrico como o *Teste de Kolmogorov-Smirnov*.

## Normal (Gaussiana)

As vezes, o pesquisar deseja que as suas va se ajustem a uma fdp *Normal($\mu$,$\sigma$)* por conta de uma série de vantagens que essa fdp permite obter dos dados. No nosso caso, usaremos a mesma va usada na fdp Gamma para verificar se uma fdp Normal pode ser aplicado a esse tipo de dados.

Estimar os parâmetros de uma fdp Normal é fácil pois necessitamos calcular apenas a média e o desvio padrão dos dados:

```r
media <- mean(dados.f$Sleep_Hours_Per_Night); round(media, 3)
```

```r
desvio <- sd(dados.f$Sleep_Hours_Per_Night); round(desvio, 3)
```

A visualização do histograma desta variável e o ajuste de uma normal (e as outras fdps Gamas estimadas anteriormente) são dada por meio do código abaixo:

```r
# 1. Estimativa dos parâmetros pelos Momentos
media_x <- mean(x)
var_x <- var(x)

shape_mom <- (media_x^2) / var_x
rate_mom <- media_x / var_x

# 2. Estimativa pela Máxima Verossimilhança
fit_gamma <- suppressWarnings(fitdistr(x, "gamma"))

shape_mle <- fit_gamma$estimate['shape']
rate_mle <- fit_gamma$estimate['rate']

# 3. Criar histograma da variável
hist(x,
     breaks = 20,
     xlim = c(1,11),
     freq = FALSE,
     col = "lightblue",
     border = "white",
     main = "Ajuste da Distribuição Gamma a 'Sleep Hours per Night'",
     xlab = "Sleep Hours per Night",
     ylab = "Frequência/Densidade de Frequência")

# 4. Sobrepor as densidades estimadas

# Sequência para traçar as densidades
x_seq <- seq(min(x)-1, max(x)+2, length.out = 500)

# Densidade Gamma pelos Momentos
dens_mom <- dgamma(x_seq, shape = shape_mom, rate = rate_mom)

# Densidade Gamma pela Verossimilhança
dens_mle <- dgamma(x_seq, shape = shape_mle, rate = rate_mle)

# Densidade da Normal pela Verossimilhança
dens_nor <- dnorm(x_seq, mean = media_x, sd = sqrt(var_x))

# Linhas de densidade
lines(x_seq, dens_mom, col = "blue", lwd = 4, lty = 2)
lines(x_seq, dens_mle, col = adjustcolor("darkorange",.75), lwd = 4, lty = 1)
lines(x_seq, dens_nor, col = adjustcolor("black",.75), lwd = 4, lty = 1)

# 5. Legenda
legend("topleft",
       legend = c("Gamma - Momentos", "Gamma - Verossimilhança", "Normal - Verossimilhança"),
       col = c("blue", "darkorange","black"),
       lwd = 4,
       lty = c(2, 1, 1),
       bty = "n")
```

Apesar de não ser a melhor distribuição para esse tipo de dado, e pela baixa assimetria (*skewness*), visualmente podemos considerar que a va `Sleep_Hours_Per_Night` possui uma distribuição aproximadamente *Normal(7.094;1.296)*.
