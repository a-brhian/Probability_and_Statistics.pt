#**Variáveis Aleatórias: Discretas e Contínuas**


No **R** temos as seguintes funções genéricas para o estudo das variáveis aleatórias:

- `dfuncao`: calcula a função de densidade no ponto *x* (discreta: função de massa, contínua: densidade de probabilidade)
- `rfuncao`: amostragem aleatória (gera valores aleatórios conforme a distribuição).
- `pfuncao`: mede a função de distribuição acumulada no ponto *x* (CDF, cumulative distribution function).

Aqui nós vamos explorar 3 funções discretas (Bernoulli, Binomial e Poison) e 2 funções contínuas (Gama e Normal).

Uma variável aleatória (va) é representada por uma letra latina maiúscula $X$ e um valor de uma amostra  é representada por uma letra latina minúscula $x$.

##**Função de Distribuição de Probabilidade Bernoulli**

Usada para modelar eventos que só podem ter dois resultados, geralmente 0 para fracasso ou 1 para sucesso.

Parâmetro:

- $p \in [0,1]$ é a probabilidade de sucesso

Função Massa de Probabilidade (fmp):

$P(X=x)=P(x)=p^x(1-p)^{1-x}, x \in \{0,1\}$

Representação:

*X $\sim$ Bernoulli(p)*

Considere uma fmp Bernoulli com $p=0.7$:

```r
# Parâmetro
p <- 0.7

# Valores possíveis da variável aleatória (0 e 1)
x <- c(0, 1)

# Probabilidades para cada valor
probabilidades <- c(1 - p, p)

# Fazer o barplot e guardar a posição das barras
bp <- barplot(probabilidades,
              main = "Distribuição de Bernoulli (0.7)",
              ylim = c(0,1),
              xlab = "Valor",
              ylab = "Probabildade",
              names.arg = x,
              col = 1)

# Adicionar os valores em cima de cada barra
text(x = bp, y = probabilidades, labels = probabilidades, pos = 3, cex = 1.5)
```

Em um estudo de simulação, considere várias amostras com $n=\{10,50,100,500,1000\}$ onde a probabilidade de acontecer um determinado evento é de $0.7$.

##**Função de Distribuição de Probabilidade Binomial**

Modela o número de sucessos em $n$ tentativas independentes de um experimento de Bernoulli, com probabilidade de sucesso $p$ em cada tentativa.

Parâmetros:

- $n$ é o número de tentativas
- $p \in [0,1]$ é a probabilidade de sucesso

fmp:

$P(X=k)=P(k)=\binom{n}{k} p^k(1-p)^{n-k}, k = 0,1,...,n$,

onde $\binom{n}{k}$ é o coeficiente binomial.

Representação:

*X $\sim$ Binomial(n,p)*

Considere uma fmp Binomial com $n = 10$ e $p=0.5$:

```r
# Parâmetros
n <- 10  # Número de tentativas
p <- 0.5  # Probabilidade de sucesso

# Calcular as probabilidades para cada número de sucessos (0 a 10)
sucessos <- 0:n  # Sucessos possíveis (de 0 a 10)
probabilidades <- dbinom(sucessos, size = n, prob = p)

# Mostrar as probabilidades
sprintf("%.3f",probabilidades)
```

O histograma dessa fmp é:

```r
# Plotar a distribuição binomial usando barplot
bp <- barplot(probabilidades,
        names.arg = sucessos,
        col = 1,
        main = "Distribuição Binomial (n=10, p=0.5)",
        xlab = "Número de Sucessos",
        ylab = "Probabilidade",
        ylim = c(0, max(probabilidades) * 1.1))  # Ajustar o limite do eixo y

# Adicionar os valores em cima de cada barra
text(x = bp, y = probabilidades, labels = sprintf("%.3f",probabilidades), pos = 3, cex = 1)
```

As vezes desejamos saber qual a probabilidade de alcançarmos até determinados números de evento. Para isso, podemos utilizar o que chamamos de *Função de Distribuição Aculmulada* (FDA):

$F(k)=P(X\leq x)=P(0)+P(1)+...+P(k)$

Em forma de gráfico pode representar a FDA da fmp anterior por:

```r
# Parâmetros
n <- 10  # Número de tentativas
p <- 0.5  # Probabilidade de sucesso

# Calcular as probabilidades acumuladas para 0 a 10 sucessos
cdf <- pbinom(0:n, size = n, prob = p)

# Plotar a CDF com traços horizontais
plot(0:n, cdf, type = "s", col = "red",
     main = "FDA da Binomial(10,0.5)",
     xlab = "Número de Sucessos", ylab = "Probabilidade Acumulada",
     lwd = 2, axes = FALSE, ylim = c(0, 1.1))

# Adicionar os valores acima de cada ponto
text(0:n, cdf, labels = sprintf("%.3f",cdf), pos = 3, col = 1)

# Adicionar eixo y
axis(2)

# Adicionar eixo x
axis(1, at = 0:n)
```

##**Função de Distribuição de Probabilidade Poison**

é uma distribuição de probabilidade discreta que modela o número de eventos que ocorrem em um intervalo fixo de tempo ou espaço, quando os eventos acontecem de forma independente e com uma taxa média constante $\lambda$.

Parâmetros:

- $\lambda$ é o taxa média de eventos por intervalo (também é a média e a variância).
- $k$ é o número de eventos

fmp:

$P(X=k)=P(k)=\frac{\lambda^k e^{-\lambda}}{k!}, k = 0,1,...$,

onde $k!=k\times(k-1)\times(k-2)...2\times 1$ é o número fatorial.

Representação:

*X $\sim$ Poison($\lambda$)*

Considere uma fmp Poison com $\lambda = 5$:

```r
# Parâmetros
lambda <- 5
k <- 0:15  # Número de eventos (de 0 a 10)

# Calcular as probabilidades de Poisson para k
probabilidades <- dpois(k, lambda)

# Plotar a distribuição de Poisson com traços horizontais
plot(k, probabilidades, type = "h", col = 1,
     main = "Distribuição de Poisson (5)",
     xlab = "Número de Eventos", ylab = "Probabilidade",
     lwd = 2, axes = FALSE, ylim = c(0, 0.25))

# Adicionar os valores acima de cada ponto
text(k, probabilidades, labels = sprintf("%.2f",probabilidades), pos = 3, col = 1)

# Adicionar eixos
axis(2)
axis(1, at = k)
```

A FDA desta distribuição seria:

```r
# Calcular as probabilidades acumuladas de Poisson (CDF)
cdf <- ppois(k, lambda)

# Plotar a CDF com traços horizontais
plot(k, cdf, type = "s", col = 1,
     ylim = c(0,1.1),
     main = "FDA da Poisson (5)",
     xlab = "Número de Eventos", ylab = "Probabilidade Acumulada",
     lwd = 2, axes = FALSE)

# Adicionar os valores acima de cada ponto
text(k, cdf, labels = sprintf("%.3f", cdf), pos = 3, col = "red")

# Adicionar eixos
axis(2)
axis(1, at = k)
```

##**Função de Distribuição de Probabilidade Normal**

A distribuição normal (ou Gaussiana) é uma distribuição de probabilidade contínua que descreve como os valores de uma variável estão distribuídos em torno da média. É uma das distribuições mais importantes em estatística, usada para modelar uma grande variedade de fenômenos naturais, como alturas, pesos e erros de medição.

Parâmetros:

- $\mu$ é a média da distribuição (pico da curva);
- $\sigma$ é o desvio padrão, que controla a largura da curva.

Função de Distribuição de Probabilidade (fdp):

### $f(x)=\frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma ^2}}, x \in \mathbb{R}$.

A curva da distribuição normal é simétrica em torno da média $\mu$ e tem forma de sino.

Representação:

*X $\sim$ Normal($\mu,\sigma$)*

Considere uma fdp Normal com $\mu = 0$ e $\sigma = 1$ conhecida como *Normal Padrão*:

```r
# Parâmetros
mu <- 0
sigma <- 1
x <- seq(-4, 4, length.out = 100)

# Plotando a função de densidade
plot(x, dnorm(x, mean = mu, sd = sigma), type = "l", col = 1, lwd=2,
     main = "Distribuição Normal (0,1)",
     xlab = "Valor", ylab = "Densidade de Probabilidade")

abline(v=mu,lty=2)
```

A FDA dessa distribuição é:

```r
# Parâmetros da normal
mu <- 0
sigma <- 1
x <- seq(-4, 4, length.out = 100)

# Plotando a FDA
plot(x, pnorm(x, mean = mu, sd = sigma), type = "l", col = 1, lwd=2,
     main = "FDA da Normal(0,1)",
     xlab = "Valor", ylab = "Probabilidade Acumulada")

abline(v=mu,lty=2)
abline(h=c(0,1),lty=3,lwd=2)
```

##**Função de Distribuição de Probabilidade Gama**

A distribuição gama é uma distribuição de probabilidade contínua que modela o tempo até a ocorrência de um número fixo de eventos em um processo de Poisson. Ela é frequentemente usada para modelar fenômenos como tempos de espera, duração de tarefas ou vida útil de dispositivos.

Parâmetros:

- $\alpha$ ou $k$ é o parâmetro de forma (número de eventos ou “picos”);
- $\beta$  é o parâmetro de escala (duração média entre eventos)

fdp:

### $f(x;\alpha,\beta)=f(x|\alpha,\beta)=\frac{x^{\alpha-1}e^{-\frac{x}{\beta}}}{\Gamma(\alpha)\beta^{\alpha}}, x \in \mathbb{R}^+$,

onde $\Gamma(\alpha)$ é a *função gama*, que generaliza o fatorial para números reais.

Representação:

*X $\sim$ Gama($\alpha,\beta$)*

Considere uma fdp Gama com $\alpha = 2$ e $\beta = 3$:

```r
# Parâmetros da distribuição gama
shape <- 2  # Parâmetro de forma
scale <- 3  # Parâmetro de escala

# Valores de x
x <- seq(0, 15, length.out = 100)

# Plotando a FDP da distribuição gama
plot(x, dgamma(x, shape = shape, scale = scale), type = "l", col = 1, lwd=2,
     main = "Gama (2,3)",
     xlab = "Valor", ylab = "Densidade de Probabilidade")
```

A FDA dessa distribuição é:

```r
# Valores de x
x <- seq(0, 20, length.out = 100)

# Plotando a CDF da distribuição gama
plot(x, pgamma(x, shape = shape, scale = scale), type = "l", col = 1, lwd =2,
     main = "FDA da Gama(2,3)",
     xlab = "Valor", ylab = "Probabilidade Acumulada")

abline(h=c(0,1),lty=3,lwd=2)
```
