Neste exercício, vamos analisar a temperatura de cocção (ato de cozinhar os alimentos) de dois tipos de alimentos: *carne bovina* e *frango*.

O objetivo é verificar se há uma diferença significativa entre as temperaturas de cocção desses dois alimentos.

Mas primeiro, vamos vamos consider que a temperatura de cocção da carne bovina tem uma média de 70°C e desvio padrão de 5°C, enquanto a temperatura de cocção do frango tem uma média de 75°C e desvio padrão de 6°C.


```r
#Vamos gerar uma semente de valores aleatório
set.seed(123)

# Temperaturas para carne bovina (média 70, desvio padrão 5)
carne_bovina <- round(rnorm(30, mean = 70, sd = 5),1)

# Temperaturas para frango (média 75, desvio padrão 6)
frango <- round(rnorm(30, mean = 75, sd = 6),1)

# Organizar os dados em um data frame
dados <- data.frame(
  Temperatura = c(carne_bovina, frango),
  Tipo_Alimento = rep(c("Carne Bovina", "Frango"), each = 30)
)

#Mostrando os dados gerados
cbind("#" = seq(1,30,1),"Carne Bovina" = carne_bovina, "Frango" = frango)
```

#Formas de visualização de dados

##Diagrama de Caule e Folha *(stem-and-leaf)*

Considere que desejaríamos verificar as informações da temperatura de cocção da carne e do frango por meio do "diagrama de caule e folha".

A função que utilizaremos para isso é `stem()`.

```r
# Organizando os dados em ordem crescente
sort(carne_bovina)

# Gráfico de caule e folha para carne bovina
cat("\nGráfico de caule e folha para carne bovina:\n")
stem(carne_bovina)
```

```r
# Organizando os dados em ordem crescente
sort(frango)

# Gráfico de caule e folha para frango
cat("\nGráfico de caule e folha para frango:\n")
stem(frango)
```

##Gráfico de pontos

O "gráfico de pontos" é uma outra forma de visualização de dados. No R ela pode ser produzida por meio da função `stripchart`.

```r
stripchart(round(frango,0),method = "stack", main = "Cocção do Frango",
           xlab = "Temperatura (°C)", ylab = "Frequência Absoluta",
           col = "darkorange", pch = 19, vertical = F,
           cex = 2, xlim = c(63, 89),
           cex.axis = 1.5,
           cex.lab = 1.6,
           cex.main = 1.8)

abline(v = seq(60, 90),
       col = adjustcolor("gray", 0.4),
       lwd = 2)
```

#Histograma (*histogram*)

O "histograma" é a forma mais comum de mostrar a frequência com que os dados aparecem em uma determinada amostra (ou população).

Neste tipo de gráfico é importante decidirmos quantos intervalos de tamanhos iguais estaremos dividindo a amostra. Existem várias metodologias para isso, e vamos começar a ver o método da *raiz quadrada*:

$$
k = \lceil \sqrt{n}  \rceil,
$$

onde $n$ é o tamanho da amostra.

```r
k <- sqrt(2)
k
```

O histograma no R é feito por meio da função `hist()`

```r
hist(frango, breaks = seq(min(frango), max(frango), length.out = k+1), main = "Cocção do Frango",
     xlab = "Temperatura (°C)", ylab = "Frequência", col = "darkorange", border = "white",
     cex.axis = 1.5, cex.lab = 1.6, cex.main = 1.8)
```

A função `hist()` do R estima o tamanho do *intervalo* por meio da *Regra de Sturges*:

$$
k = \lceil \log_2(n) + 1 \rceil
$$

```r
k <- ceiling(log2(length(frango))+1)
k
```

```r
intervalos <- seq(min(frango), max(frango), length.out = k+1)

h <- hist(frango, breaks = intervalos, main = "Cocção do Frango",
     xlab = "Temperatura (°C)", ylab = "Frequência", col = "darkorange", border = "white",
     cex.axis = 1.5, cex.lab = 1.6, cex.main = 1.8, xaxt = "n", yaxt = "n")

axis(2, at = seq(0, max(h$counts), by = 1), cex.axis = 1.5)

axis(1, at = intervalos, cex.axis = 1.5, labels = sprintf("%.1f",intervalos))
```

Mas pela frente iremos mostrar as medidas de localização em um histograma e apresentaremos uma outra forma de visualizar os dados por meio do *boxplot*.


#Medidas de localização (posição ou tendência central)

##Média Amostral

A média é calculada por meio da função `mean()`.

```r
mean(frango)
```

##Mediana

A mediana é calculada por meio da função `median` ou `quantile()`.

```r
median(frango)
cat("\n")
quantile(frango,probs=0.5)
```

##Quartis

Os quartis são calculados por meio da função `quantile()`.

```r
cat("1° quartil\n")
quantile(frango,probs = 0.25)
cat("\n2° quartil ou mediana\n")
quantile(frango,probs = 0.5)
cat("\n3° quartil\n")
quantile(frango,probs = 0.75)
```

Assim, todas essas medidas de localização poderiam ser representadas dentro do histograma:

```r
#Histograma
h <- hist(frango, breaks = intervalos, main = "Cocção do Frango",
     xlab = "Temperatura (°C)", ylab = "Frequência", col = "darkorange", border = "white",
     cex.axis = 1.5, cex.lab = 1.6, cex.main = 1.8, xaxt = "n", yaxt = "n")

#Configuração dos eixos
axis(2, at = seq(0, max(h$counts), by = 1), cex.axis = 1.5)
axis(1, at = intervalos, cex.axis = 1.5, labels = sprintf("%.1f",intervalos))

#Legenda
legend("topright", legend = c("Média", "Quartis"), text.col = c("blue", "skyblue"),
       col = c("blue", "skyblue"), lty = c(NA, NA), lwd = 5, cex = 2, bty = "n")

#Linhas verticais
abline(v=mean(frango),col=adjustcolor("blue",.9),lty=2,lwd=5)
abline(v=quantile(frango,probs = c(.25,.5,.75)),col=adjustcolor("skyblue",.9),lty=1,lwd=5)
```

#Medidas de dispersão

##Desvio Padrão

O desvio padrão amostral é calculado por meio da função `sd()`.

```r
sd(frango)
```

A variância amostral é dada pelo quadrado do desvio padrão amostral ou é calculado pela função `var()`.

```r
var(frango)
cat("\n")
(sd(frango))^2
```

##Desvio Interquartílico

O desvio interquartílico é feito pela diferença entre o 3º e o 1º quartil da amostra.

```r
desvio.int <- quantile(frango,probs=.75) - quantile(frango,probs=.25)
desvio.int
```

#Diagrama de Caixa ou Gráfico de Caixa (*boxplot*)

O boxplot é feito por meio da função `boxplot()`:

```r
boxplot(frango,
        horizontal=T,
        main = "Cocção do Frango",
        ylab = "Temperatura (°C)",
        col = "darkorange",
        border = "black",
        cex.axis = 1.4,
        cex.lab = 1.6,
        cex.main = 1.8)
```

Agora vamos adicionar ao *boxplot* as medidas de tendência central:

```r
bp <- boxplot(frango,
        horizontal=T,
        main = "Cocção do Frango",
        ylab = "Temperatura (°C)",
        col = "darkorange",
        border = "black",
        cex.axis = 1.4,
        cex.lab = 1.6,
        cex.main = 1.8,
        xaxt = "n")

legend("topright", legend = c("Média", "Quartis"), text.col = c("blue", "skyblue"),
       col = c("blue", "skyblue"), lty = c(NA, NA), lwd = 5, cex = 2, bty = "n")

valores <- bp$stats[,1]
axis(1, at = valores, labels = sprintf("%.1f", valores), cex.axis = 1.5)

abline(v=mean(frango),col=adjustcolor("blue",.9),lty=2,lwd=5)
abline(v=quantile(frango,probs = c(.25,.5,.75)),col=adjustcolor("skyblue",.7),lty=1,lwd=5)
```

Vamos verificar como que o *boxplot* ficaria sobreposto a um histograma:

```r
#Histograma
h <- hist(frango, breaks = intervalos, main = "Cocção do Frango",
     xlab = "Temperatura (°C)", ylab = "Frequência", col = "darkorange", border = "white",
     cex.axis = 1.5, cex.lab = 1.6, cex.main = 1.8, xaxt = "n", yaxt = "n")

axis(2, at = seq(0, max(h$counts), by = 1), cex.axis = 1.5)
axis(1, at = intervalos, cex.axis = 1.5, labels = sprintf("%.1f",intervalos))

legend("topright", legend = c("Média", "Quartis"), text.col = c("blue", "skyblue"),
       col = c("blue", "skyblue"), lty = c(NA, NA), lwd = 5, cex = 2, bty = "n")

abline(v=mean(frango),col=adjustcolor("blue",.9),lty=2,lwd=5)
abline(v=quantile(frango,probs = c(.25,.5,.75)),col=adjustcolor("skyblue",.9),lty=1,lwd=5)

#Comando para sobreposição de gráficos
par(new = TRUE)

#Boxplot
bp <- boxplot(frango,
        horizontal=T,
        main = "",
        ylab = "",
        col = adjustcolor("darkorange",.5),
        border = "black",
        cex.axis = 1.4,
        cex.lab = 1.6,
        cex.main = 1.8,
        xaxt = "n")

legend("topright", legend = c("Média", "Quartis"), text.col = c("blue", "skyblue"),
       col = c("blue", "skyblue"), lty = c(NA, NA), lwd = 5, cex = 2, bty = "n")

abline(v=mean(frango),col=adjustcolor("blue",.9),lty=2,lwd=5)
abline(v=quantile(frango,probs = c(.25,.5,.75)),col=adjustcolor("skyblue",.7),lty=1,lwd=5)
```
